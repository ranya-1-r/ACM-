<h3>Transformer les variables quantitatives en variables catégoriales</h3>

```python
# Transformer la variable age en variable catégorielle en créant des groupes

# Définir les limites des groupes d'âge
age_bins = [18, 30, 40, 50, 60, 100]

# Définir les labels pour les groupes d'âge
age_labels = ['[18-29]', '[30-39]', '[40-49]', '[50-59]', '[+59]']

# Créer une nouvelle colonne 'Age_group' en utilisant la fonction cut()
mall_data['Age_group'] = pd.cut(mall_data['Age'], bins=age_bins, labels=age_labels, right=False)


# Transformer la variable Annual Income en variable catégorielle en créant des groupes

Income_bins = [15, 38, 61, 84, 107, 138]

Income_labels = ['[15-37]','[38-60]','[61-83]','[84-106]','[107-138]']

mall_data['Income_grps'] = pd.cut(mall_data['Annual Income (k$)'], bins=Income_bins, labels=Income_labels, right=False)

# Transformer la variable Spending Score en variable catégorielle en créant des groupes

Spending_bins = [1, 21, 41, 61, 81,101]

Spending_labels = ['[1-20]','[21-40]','[41-60]','[61-80]','[81-100]']

mall_data['Spending_grps'] = pd.cut(mall_data['Spending Score (1-100)'], bins=Spending_bins, labels=Spending_labels, right=False)

mall_data.head()
```
